### Exercício 1


Ler 5 notas e informar a média usando loop for


```python
soma_notas = 0

for i in range(5):
  nota = float(input(f"Digite a {i+1}ª nota: "))
  soma_notas += nota

media = soma_notas / 5

print(f"A média das notas é: {media}.")
```

### Exercício 2
Ler 5 notas e informar a média usando loop while

```python
soma_notas = 0
contador_notas = 0

while contador_notas < 5:
  nota = float(input(f"Digite a {i+1}ª nota: "))
  soma_notas += nota
  contador_notas += 1

media = soma_notas / 5

print(f"A média das notas é: {media}.")
```

### Exercício 3

Imprimir a tabuada do número 3 (3 x 1 = 1 - 3 x 10 = 30) usando loop for

```python
numero = 3

for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i} = {resultado}")
```

### Exercício 4

Imprimir a tabuada do número 3 (3 x 1 = 1 - 3 x 10 = 30) usando loop while

```python
numero = 3
i = 1

while i <= 10:
    resultado = numero * i
    print(f"{numero} x {i} = {resultado}")
    i += 1
```
