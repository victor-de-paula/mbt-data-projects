## **Operadores aritméticos**

Os operadores aritméticos são usados com valores numéricos para realizar operações matemáticas.

Operador | Nome
-------- | -------
+        | Adição
-        | Subtração
*        | Multiplicação
**       | Exponenciação
/        | Divisão
//       | Divisão Inteira (Quociente)
%        | Módulo (Resto da divisão)

````python
# Adição

numero_1 = 9
numero_2 = 2

soma = numero_1 + numero_2

print(f'A soma dos números {numero_1} + {numero_2} é igual a {soma}')
````

````python
numero_1 = int(input('Digite um número: '))
numero_2 = int(input('Digite outro número: '))
````

````python
soma = numero_1 + numero_2

print(f'A soma dos números {numero_1} + {numero_2} é igual a {soma}')
````

````python
# Subtração

subtracao = numero_1 - numero_2

print(f'A subtração dos números {numero_1} - {numero_2} é igual a {subtracao}')
````

````python
# Multiplicação

multiplicacao = numero_1 * numero_2

print(f'A multiplicação dos números {numero_1} * {numero_2} é igual a {multiplicacao}')
````

````python
# Exponenciação

exponenciacao = numero_1 ** numero_2

print(f'A exponenciação dos números {numero_1} ** {numero_2} é igual a {exponenciacao}')
````

````python
# Divisão

divisao = numero_1 / numero_2

print(f'A divisão dos números {numero_1} / {numero_2} é igual a {divisao}')
````

````python
# Divisão inteira

divisao_inteira = numero_1 // numero_2

print(f'A divisão inteira dos números {numero_1} // {numero_2} é igual a {divisao_inteira}')
````

````python
# Resto da divisão

resto_divisao = numero_1 % numero_2

print(f'O resto da divisão dos números {numero_1} % {numero_2} é igual a {resto_divisao}') # Se resultado 0, então par; do contrário, ímpar
````

````python
soma = 10 + 2
````

````python
if c > d:
  print(f'{c} é maior que {d}.')

elif d > c:
  print(f'{d} é maior que {c}.')

else: # Else é sempre a condição final e não preciso especificar a condição
  print(f'Os valores {c} e {d} são iguais.')
````

# **Aula 06 - Estrutura de Dados - Listas**

___
Uma lista (list) em Python é uma coleção de valores, separados por vírgula e dentro de colchetes [].

**Métodos da lista**
* **append(x)**: Adiciona um item ao fim da lista;
* **extend(iterable)**: Prolonga a lista, adicionando no fim todos os elementos passados como parâmetro;
* **insert(i, x)**: Insere um item em uma dada posição. O primeiro argumento é o índice do elemento antes do qual será feita a inserção;
* **remove(x)**: Remove o primeiro intem encontrado na lista cujo valor é igual a x;
* **pop([i])**: Remove um item em uma dada posição na lista e o retorna. Se nenhum índice é especificado, a pop() remove e devolve o último item da lista;
* **clear(x)**: Remove todos os itens de uma lista;
* **index(x)**: Devolve o índice do primeiro item cujo valor é igual a x;
* **count(x)**: Devolve o número de vezes em que x aparece na lista;
* **sort(*, key=None, reverse=False)**: Ordena os itens na lista (os argumentos podem ser usados para personalizar a ordenação).

# **Aula 08 - Estrutura de Dados - Dicionários**

___
Um dicionário é um tipo de estrutura de dados do tipo coleção, ou seja, um objeto que contém mais de um valor.

Diferentemente das listas e tuplas em que os elementos são acessados através de uma posição ou índice, nos dicionários o acesso às informações ocorrem através de chaves. Além disso, os elementos de um dicionário são armazenados de forma não ordenada.

**Métodos dos Dicionários**
* **get(key[,default])**: Retorna o valor da chave se a chave estiver no dicionário, senão default.
* **keys()**: Retorna as chaves do dicionário.
* **values()**: Retorna os valores do dicionário.
* **items()**: Retorna uma nova visão dos itens do dicionário (pares).
* **pop(key[,default])**: Se a chave estiver no dicionário, remova-a e retorne seu valor, senão retorne default.
* **popitem()**: Remove o último elemento do dicionário.
* **clear()**: Remova todos os itens do dicionário.
* **update()**: Atualize o dicionário com os pares chave/valor.

## **Criando dicionários**

# **Aula 10 - Estruturas de Repetição - WHILE**

___
O **while** é uma estrutura de repetição que permite executar um determinado bloco de código enquanto uma condição for verdadeira.

![](https://drive.google.com/file/d/15d7FC31HrRMicuuRQQo4HNjkS1Z4skmu/view?usp=sharing)while.png

````python
contador = 0
while contador < 5:
  print('Olá, mundo!')

  contador = contador + 1 # Atenção a indentação

# contador = contador + 1 -> Se executo sem indentar, não respeita a condição de repetição
````

````python
n = 1
soma = 0

while n != 0:
  numero = int(input(f'Digite um número: '))

  soma = soma + numero

  if numero == 0:
    n = 0
````

````python
soma = 0
lista = []

while True:
  numero = int(input(f'Digite um número: '))

  lista.append(numero)

  soma = soma + numero

  if numero == 0:
    break
````

````python
print(lista)
````

````python
def soma_media(a, b):
  soma = a + b
  media = (a + b) / 2

  return soma, media
````

````python
resultado_geral = soma_media(100, 20)
````

````python
print(resultado_geral)
````

````python
print(type(resultado_geral))
````

````python
soma, media = soma_media(100, 20)
````
