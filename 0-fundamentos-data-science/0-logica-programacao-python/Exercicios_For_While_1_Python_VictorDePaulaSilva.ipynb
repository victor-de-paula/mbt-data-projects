#### Exercício 1
Apresentar o quadrado dos números inteiros de 15 a 200

```python
for i in range(15, 201):
    quadrado = i ** 2
    print(f"O quadrado de {i} é {quadrado}")
```

#### Exercício 2

Apresentar o total da soma obtida dos cem primeiros números inteiros
(1+2+3+4+5+6+7+...+97+98+99+100)

```python
soma = 0

for i in range(1, 101):
  soma += i

print(f"O total da soma dos cem primeiros números inteiros é: {soma}.")
```

#### Exercício 3
Escrever um algoritmo que lê 5 valores para a variável A, um de cada vez,
e contar quantos destes valores são negativos. Após a leitura, o programa deve mostrar
a quantidade de números negativos

```python
negativos = 0

for i in range(5):
  valor = float(input(f"Digite o {i+1}º valor: "))
  if valor < 0:
    negativos += 1

print(f"A quantidade de números negativos é: {negativos}.")
```

#### Exercício 4
Ler duas notas e informar a média. Se o valor digitado for menor do que 0 (zero)
ou maior do que 10 (dez), o usuário deve digitar a nota novamente

```python
def ler_nota():
  while True:
    nota = float(input("Digite uma nota (entre 0 e 10): "))
    if 0 <= nota <= 10:
      return nota
    else:
      print("Por favor, digite uma nota válida entre 0 e 10.")

nota1 = ler_nota()
nota2 = ler_nota()

media = (nota1 + nota2) / 2

print(f"A média das notas informadas é: {media}.")
```
