##### Exercício 8  -  Programa Python para trocar os primeiros e últimos elementos em uma lista

Dada uma lista, escreva um programa Python para trocar o primeiro e o último elemento da lista.


Exemplos:
```
Input : [12, 35, 9, 56, 24],
Output : [24, 35, 9, 56, 12],

Input : [1, 2, 3],
Output : [3, 2, 1]
```



````python
lista1 = [12, 35, 9, 56, 24]
lista2 = [1, 2, 3]
````

````python
def trocar_elementos(lista):
  if len(lista) < 2:
    return

  lista[0], lista[-1] = lista[-1], lista[0]

  return lista
````

````python
print("Input:", lista1)
print("Output:", trocar_elementos(lista1))

print("Input:", lista2)
print("Output:", trocar_elementos(lista2))
````
