##### Exercício 1 - Calcule a soma dos elementos da  tupla 'my_tuple'


```python
my_tuple = 1, 2, 3, 4, 5
```

```python
soma_tupla = sum(my_tuple)
```

##### Exercício 3 - Encontrar o erro no código abaixo

```python
my_tuple = 1, 2, 3, 4, 5
my_tuple.remove(1)
print(my_tuple)
```

```python
# As tuplas são imutáveis. Assim, não há um comando nativo para remoção de um elemento.
# A alternativa, neste caso, é converter a lista em uma tupla, remover o elemento e converter de volta como tupla.

my_tuple = 1, 2, 3, 4, 5
remover = 1

new_tuple = tuple(x for x in my_tuple if x != remover)

print(f"Tupla original: {my_tuple}.")
print(f"Nova tupla: {new_tuple}.")
```
